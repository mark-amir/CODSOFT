```python
colors = ["red", "blue" , "black"]
species = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
```

```python
for i in range(3):
    x= df[df['species']==species[i]]
    plt.scatter(x['sepal_length'],x['sepal_width'],c= colors[i],label= species[i])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
```

```python
for i in range(3):
    x= df[df['species']==species[i]]
    plt.scatter(x['petal_length'],x['petal_width'],c= colors[i],label= species[i])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend()
```

```python
for i in range(3):
    x= df[df['species']==species[i]]
    plt.scatter(x['sepal_length'],x['petal_length'],c= colors[i],label= species[i])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend()
```

```python
for i in range(3):
    x= df[df['species']==species[i]]
    plt.scatter(x['sepal_width'],x['petal_width'],c= colors[i],label= species[i])
plt.xlabel("Sepal Width")
plt.ylabel("Petal Width")
plt.legend()
```

```python
X = df.drop(columns = "species")
Y = df["species"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2) 
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_pred = model.predict(X_train)
X_train_pred
```

```python
train_data_accuracy = accuracy_score(Y_train,X_train_pred)
```

```python
print("Accuracy of data:", train_data_accuracy) 
```

```python
X_test_pred = model.predict(X_test)
X_test_pred
```

```python
test_data_accuracy = accuracy_score(Y_test,X_test_pred)
```

```python
print("Accuracy of data:", test_data_accuracy) 
```

```python
conf_matr = classification_report(Y_test, X_test_pred)
print("Confusion matrix:", conf_matr)
```

```python
class_report = classification_report(Y_test, X_test_pred, target_names=['setosa', 'versicolor', 'virginica'])
print("Classification report:" , class_report)
```
